## Aplicamos transformaciones a una imagen
Con este paso verificamos que los resultados en una imagen sean los deseados y  realizar las modificaciones necesarias.

Mostramos una imagen

```python
for line in weights_data:
  img_name, weight = line.strip().split(',')

print(img_name)
print(train_folder)
img_path = os.path.join(train_folder, img_name)
print(img_path)

# Leer imagen con OpenCV
img = cv2.imread(img_path)

if img is None:
    print(f"No se pudo leer la imagen {img_name}.")
else:
    print(f"Se pudo leer la imagen {img_name}.")

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Aplicamos una rotaci√≥n a la imagen

```python
# Dimensiones de la imagen original
h, w = img.shape[:2]
center = (w // 2, h // 2)

# Matriz de rotaci√≥n
angle = 15
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
# Aplicar la rotaci√≥n
rotated_img = cv2.warpAffine(img, rotation_matrix, (w, h))

# Combinar las im√°genes horizontalmente en un solo lienzo
combined = np.hstack((img, rotated_img))  # np.vstack para apilar verticalmente

# Mostrar las im√°genes combinadas
cv2_imshow(combined)
```

Guardamos la imagen rotada y agregamos el peso asociado a la imagen de origen

```python
rot_img_name = f"ROT_{img_name}"
folder = '/content/drive/MyDrive/PRUEBA' # esta carpeta es de prueba
rot_img_path = os.path.join(folder, rot_img_name)
print(rot_img_path)
```

```python
# Subir la imagen a la carpeta
cv2.imwrite(rot_img_path, rotated_img)
```

```python
# Registrar en el archivo de pesos de ROT_TRAIN
rot_weights_txt = '/content/drive/MyDrive/PRUEBA/test.txt'
# Crear archivos txt si no existen
with open(rot_weights_txt, 'w') as file:
    pass

with open(rot_weights_txt, 'a') as rot_wf:
  rot_wf.write(f"{rot_img_name},{weight}\n")
```

Aplicamos el efecto Blur a una imagen

```python
# Aplicar efecto blur (GaussianBlur)
blurred_img1 = cv2.GaussianBlur(img, (7, 7), 0.5) # Kernel 5x5 y sigma=1 para desenfoque suave
blurred_img2 = cv2.GaussianBlur(img, (7, 7), 1)

# Mostrar
# Combinar las im√°genes horizontalmente en un solo lienzo
combined = np.hstack((img,blurred_img1, blurred_img2))  # np.vstack para apilar verticalmente

# Mostrar las im√°genes combinadas
cv2_imshow(combined)
```

Comparaci√≥n de ambos efectos

```python
if len(img.shape) == 3:  # Verificar si tiene 3 canales (color)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convertir a RGB
else:
    print("La imagen original est√° en escala de grises")
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convertir a RGB
rotated_img = cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB)
blurred_img1 = cv2.cvtColor(blurred_img1, cv2.COLOR_BGR2RGB)


# Configurar la figura
plt.figure(figsize=(20, 15))

# Imagen original
plt.subplot(131)
plt.imshow(img)
plt.title("Imagen Original", fontsize=50)
plt.axis("off")

# M√°scara
plt.subplot(132)
plt.imshow(rotated_img)
plt.title("Rotaci√≥n 15¬∞", fontsize=50)
plt.axis("off")

# Imagen con Unsharp Mask
plt.subplot(133)
plt.imshow(blurred_img1, vmin=0, vmax=1)
plt.title("Gaussian Blur", fontsize=50)
plt.axis("off")

# Mostrar las im√°genes
plt.tight_layout()
plt.show()
```

## Para el dataset
Aplicamos las transformaciones (rotar y efecto blur) al conjunto de imagenes del dataset

```python
# A√±adir registros a los archivos de pesos para nuevas carpetas
with open(rot_weights_file, 'a') as rot_wf, open(blur_weights_file, 'a') as blur_wf:
    for line in weights_data:
        img_name, weight = line.strip().split(',')
        img_path = os.path.join(train_folder, img_name)
        print(img_path)

        # Leer imagen con OpenCV
        img = cv2.imread(img_path)

        if img is None:
            print(f"No se pudo leer la imagen {img_name}.")
            continue
        else:
            print(f"Se pudo leer la imagen {img_name}.")

        # Rotar la imagen 15 grados
        h, w = img.shape[:2]
        center = (w // 2, h // 2)
        rotation_matrix = cv2.getRotationMatrix2D(center, 15, 1.0)
        rotated_img = cv2.warpAffine(img, rotation_matrix, (w, h))
        # Nombrar cada imagen
        rot_img_name = f"ROT_{img_name}"
        rot_img_path = os.path.join(rot_train_folder, rot_img_name)
        cv2.imwrite(rot_img_path, rotated_img)

        # Registrar en el archivo de pesos de ROT_TRAIN
        rot_wf.write(f"{rot_img_name},{weight}\n")

        # Aplicar efecto blur (GaussianBlur)
        blurred_img = cv2.GaussianBlur(img, (5, 5), 1)  # Kernel 5x5 y sigma=1 para desenfoque suave
        blur_img_name = f"BLUR_{img_name}"
        blur_img_path = os.path.join(blur_train_folder, blur_img_name)
        cv2.imwrite(blur_img_path, blurred_img)

        # Registrar en el archivo de pesos de BLUR_TRAIN
        blur_wf.write(f"{blur_img_name},{weight}\n")

print("Proceso completado.")
```
